## Running data_analysis.py

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('provided_data.csv')

# Display the first 5 rows
print(df.head())

# Display basic information about the dataset
print(df.info())

# Calculate and print summary statistics
print(df.describe())

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df.iloc[:, 0], df.iloc[:, 1])
plt.xlabel('Frame Number')
plt.ylabel('Value')
plt.title('Second Column vs Frame Number')
plt.grid(True)
plt.savefig('plot.png')
plt.show()
```

## Modifications

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('provided_data.csv')

# Display the first 5 rows
print(df.head())

# Display basic information about the dataset
print(df.info())

# Calculate and print summary statistics
print(df.describe())

# Plotting
plt.figure(figsize=(10, 6))
for i in range(1, 6):
    plt.plot(df.iloc[:, 0], df.iloc[:, i])
    plt.xlabel('Frame Number')
    plt.ylabel('Value')
    plt.title(f'Column {i} vs Frame Number')
    plt.grid(True)
    plt.savefig(f"images/plot{i}.png")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('provided_data.csv')
plt.figure(figsize=(10, 6))
plt.xlabel("Frame Number")
plt.ylabel("Value")

plt.title("Line Plot")
plt.plot(df.iloc[:, 0], df.iloc[:, 1])
plt.savefig("images/line-plot.png")

plt.title("Scatter Plot")
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.savefig("images/scatter-plot.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("provided_data.csv")

plt.figure(figsize=(10, 6))

for i in range(1, 5):
    plt.plot(df.iloc[:, 0], df.iloc[:, i], label=f'Column {i}')

plt.xlabel('Frame Number')
plt.ylabel('Value')
plt.title('Multiple Columns vs Frame Number')
plt.legend()
plt.savefig("images/multiple-plot.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("provided_data.csv")

plt.figure(figsize=(10, 6))

plt.hist2d(df.iloc[:, 0], df.iloc[:, 1])

plt.xlabel('Frame Number')
plt.ylabel('Value')
plt.title('Histogram of Column 1 vs Frame Number')
plt.savefig("images/histogram.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("provided_data.csv")

fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].hist2d(df.iloc[:, 0], df.iloc[:, 1])
axes[0, 1].hist2d(df.iloc[:, 0], df.iloc[:, 2])
axes[1, 0].hist2d(df.iloc[:, 0], df.iloc[:, 3])
axes[1, 1].hist2d(df.iloc[:, 0], df.iloc[:, 4])

plt.tight_layout()
plt.savefig("images/sub-plots.png")
```
